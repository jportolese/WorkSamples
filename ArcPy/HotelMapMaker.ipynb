```python
import arcpy,os

arcpy.env.overwriteOutput = True

# Project object
aprx = arcpy.mp.ArcGISProject(
    r"C:\GIS\HotelMaps\HotelMaps.aprx")

# Map objects
mainMap = aprx.listMaps("MainMap")[0]
overviewMap = aprx.listMaps("OverviewMap")[0]

# Layer objects
HotelsLayer = mainMap.listLayers("HotelMarketsFinal")[0]
SubsLayer = mainMap.listLayers("HotelSubmarketsFinal")[0]

# Layout object
lyt = aprx.listLayouts()[0]

# Map Frame objects
mainMapFrame = lyt.listElements('MAPFRAME_ELEMENT',"MainMap Frame")[0]
overviewMapFrame = lyt.listElements('MAPFRAME_ELEMENT',"OverviewMap Frame")[0]


MarketsSortedByNameList = sorted([row[0] for row in arcpy.da.SearchCursor(
    HotelsLayer,"Marketname")])

for pageCount, MarketName in enumerate(MarketsSortedByNameList[:65]):  

            HotelsLayer.definitionQuery = "Marketname = '{0}'".format(MarketName)
            SubsLayer.definitionQuery = "Marketname = '{0}'".format(MarketName)
            selMarketExtent = mainMapFrame.getLayerExtent(HotelsLayer)
            mainMapFrame.camera.setExtent(selMarketExtent)
            mainMapFrame.camera.scale = mainMapFrame.camera.scale * 1.05
            print(MarketName)

            # Title Text Object
            titleText = lyt.listElements("TEXT_ELEMENT")[0]
            titleText.text = MarketName

            # Export PDF for this country's page
            if MarketName == 'Washington, DC/MD/VA' :
                lyt.exportToPDF(r"C:\GIS\HotelMaps\PDFs\Washington, DC.pdf")
            else: # MarketName <> 'Washington, DC/MD/VA':
                lyt.exportToPDF(r"C:\GIS\HotelMaps\PDFs\{0}.pdf".format(MarketName))

# Delete project object
del aprx
```
